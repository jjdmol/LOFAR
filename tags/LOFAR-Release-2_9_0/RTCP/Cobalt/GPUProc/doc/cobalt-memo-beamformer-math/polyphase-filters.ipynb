# Polyphase filters and their inverses

A polyphase filter is a way to improve the frequency response of a channel in a filter bank by extending the length of the timeseries used in computing the value of the channel. Effectively, if one wants a channel response that is close to a top hat (the signal that is analyzed is convolved with this top-hat function in frequency space), one should multiply the time series data with the fourier transform of this top hat: the sinc function.

Suppose we have a time series $f(t_m)$ for which we want to compute a dynamic spectrum of $N$ channels $g(\nu_n)$. We furthermore assume equidistant Nyquist sampling, so $t_m = t_c+m\Delta t$, $\nu_n = \nu_\mathrm{c} + \left(n-\left\lfloor \frac{N}{2}\right\rfloor\right)\Delta\nu$, and $\Delta \nu = {N\Delta t}^{-1}$.

If we want the top-hat function to be exactly $\Delta\nu$ wide, the weighting function in the time domain should be

\begin{eqnarray}
w(t) & = & \Delta\nu^{-1}\int_{-\frac{1}{2}\Delta\nu}^{+\frac{1}{2}\Delta\nu} \cos(2\pi\nu t)\ \mathrm{d}\nu\\
w(t) & = & (2\pi\Delta\nu 1)^{-1}\left[\sin(2\pi\nu t)\right]^{+\frac{1}{2}\Delta\nu}_{-\frac{1}{2}\Delta\nu}\\
w(t) & = & \frac{\sin(\pi\Delta\nu t)}{\pi\Delta\nu t}\\
w_m  & = & \frac{\sin(\pi\Delta\nu m \Delta t)}{\pi\Delta\nu m \Delta t}\\
w_m  & = & \frac{\sin(\pi \frac{m}{N})}{\pi \frac{m}{N}}
\end{eqnarray}



```python
%pylab inline
@vectorize
def sinc_weights(sample_m, num_chan):
    if sample_m == 0:
        return 1.0
    else:
        arg = pi*sample_m/float(num_chan)
        return sin(arg)/(arg)
```

```python
taps=16
plot(arange(-taps*64/2, taps*64/2+1), sinc_weights(arange(-taps*64/2, taps*64/2+1), 64))
grid()
```

```python
plot(abs(fft.fftshift(fft.fft(kaiser(taps*64 +1,beta=4.0)*sinc_weights(arange(-taps*64/2, taps*64/2+1), 64)))))
xlim(500,525)
grid()
```
